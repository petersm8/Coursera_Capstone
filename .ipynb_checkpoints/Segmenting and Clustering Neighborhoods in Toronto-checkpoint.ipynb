Necessary imports for this project

```python
import pandas as pd 
from bs4 import BeautifulSoup
import requests
```

url of the scraping and get s the data with BeautifulSoup

```python
url ='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
req = requests.get(url).text
soup = BeautifulSoup(req,'lxml')
```

makes and empty dataframe and gets the table from the webpage

```python
df = pd.DataFrame()
My_table = soup.find('table') 
```

iterates through each row and column and appends the data to the df

```python
for tr in My_table.find_all('tr'):
    for tds in tr.find_all('td'):
        row1=""
        row2 = []
        for bs in tds.find_all('b'):
            row1 = bs.text # This is the postalCode
        for brs in tds.find_all('a'):
            row2.append(brs.get('title')) # This is the other data
        if(row2 != []): # If the data is not Not Assigned
            if(row2[1:] == []): # Determines if the bourough and Neighborhood will be the same
                em = row2[0]
            else:
                em = ', '
                em = em.join(row2[1:])
            # Makes new data frame from this cells data
            data = {'PostalCode': [row1],'Borough':[row2[0]],'Neighborhood':[em]}
            da = pd.DataFrame(data)
            # Appends the data to the overall dataframe
            df = df.append(da)
```
